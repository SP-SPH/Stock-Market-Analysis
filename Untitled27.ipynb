```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("/content/stocks.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.plot(x='Date', y='Close', title='Stock Closing Price Over Time', figsize=(10, 5))
```

```python
df.head(2)
```

```python
import matplotlib.pyplot as plt

df['price range'] = df['High'] - df['Low']

plt.figure(figsize=(10, 5))

plt.bar(df['Date'], df['price range'])

plt.title('Stock Market Range Over Time')
plt.xlabel('Date')
plt.ylabel('Price Range')


plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
df.info()
```

```python
df.plot(x='Date', y='Volume', kind='line', title='Daily Trading Volume', figsize=(12, 4))
```

```python
df['MA7'] = df['Close'].rolling(window=7).mean()
df['MA14'] = df['Close'].rolling(window=14).mean()
df.plot(x='Date', y=['Close', 'MA7', 'MA14'], title='Moving Averages', figsize=(10, 5))
```

```python
df.info()
```

```python
total_volumeby_Ticker = df.groupby('Ticker')['Volume']
total_volumeby_Ticker.plot(kind='bar',figsize=(12,5),color='blue',title='total_volumeby_Ticker')
plt.xlabel("Ticker")
plt.ylabel("total volume")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Close'], kde=True, color='orange')
plt.title('Distribution of Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
